```python
#Plota os dados originais

plt.figure(figsize=(10,5))
plt.plot(x,y,'o', label='Dados Originais')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()
```

```python
# Fazer regressão usando a função LinearRegression do sklearn

x,y = x.reshape(-1,1), y.reshape(-1,1) # reshape(-1,1) transforma o vetor em uma matriz  coluna

```

```python
# Criação e treinamento do modelo

lr = LinearRegression() # Cria o objeto de regressão linear
lr.fit(x,y); # Treina o modelo
```

```python
print("Y = {:.4f}X + {:.4f} ".format(lr.coef_[0][0],lr.intercept_[0])) # Imprime os coeficientes da regressão
```

```python
# predição

y_pred = lr.predict(x) # predição dos valores de y

# Verificação de acurácia

score = lr.score(x,y) # score do modelo
print("Score do modelo: {:.4f}".format(score))

```

```python
# Plota os dados originais e a reta de regressão

plt.figure(figsize=(10,5))
plt.plot(x,y,'o', label='Dados Originais')
plt.plot(x,y_pred,'r', label='Reta de Regressão : acurácia = {:.4f}'.format(score))
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.title('Regressão Linear')
plt.grid()
plt.show()
```

```python
# Aprofundamento sobre o score

plt.figure(figsize=(10,5))
plt.plot(x,y,'o', label='Dados Originais')
plt.plot(x,y_pred,'r', label='Reta de Regressão : acurácia = {:.4f}'.format(score))
plt.hlines(y.mean(),x.min(),x.max(),linestyles="dashed", label='Média de y',colors='black')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.title('Regressão Linear')
plt.grid()
plt.show()
```

```python
def MSE(y,y_pred):
    return np.sum((y-y_pred)**2)/len(y)

def R2(y,y_pred):
    return 1 - (np.sum((y-y_pred)**2)/np.sum((y-y.mean())**2))
```

```python
print("MSE = {:.4f}".format(MSE(y,y_pred)))
print("MSE sklearn = {:.4f}".format(np.mean((y-y_pred)**2)))
print("R2 = {:.4f}".format(R2(y,y_pred)))
print("R2 sklearn = {:.4f}".format(lr.score(x,y)))
```
